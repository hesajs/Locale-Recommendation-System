```python
data = ImageDataBunch.from_folder(
            path,
            train='.',
            valid_pct=0.2,
            ds_tfms=get_transforms(),
            size=224,
            num_workers=4,
            ).normalize(imagenet_stats)
```

```python
model.fit_one_cycle(10)
```
